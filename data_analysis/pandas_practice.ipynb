## Pandas 로 조작해보기
- 우선, 현재 엑셀 형태 보기
![real_excel](../images/real_excel.png)
- 엑셀의 **세번째 행**을 header 로 설정하고, column 이름을 재정의하자
- 엑셀의 **첫번째 열**은 필요없는 내용이므로 버리자.

```python
import pandas as pd


filename = "../data/등록인구_20250403110803.xlsx"
original = pd.read_excel(
    filename, sheet_name=0, header=[2]
)  # 1번째 열을 3번째 행을 header 로
original
```

```python
df = original.drop(0).drop(columns=["동별(1)"])  # 필요없는 데이터 제거
df.head()
```

```python
# column 명을 각 column 이 어떤 의미인지 명확하도록 이름 재정의
df = df.rename(
    columns={
        "동별(2)": "동",
        "소계": "세대",
        "소계.1": "전체 수",
        "남자": "총 남자 수",
        "여자": "총 여자 수",
        "소계.2": "총 한국인 수",
        "남자.1": "총 한국인 남자 수",
        "여자.1": "총 한국인 여자 수",
        "소계.3": "총 외국인 수",
        "남자.2": "총 외국인 남자 수",
        "여자.2": "총 외국인 여자 수",
        "소계.4": "세대 당 평균 인구 수",
        "소계.5": "총 65세 이상 고령자 수",
    }
)
print("모든 컬럼명:", df.columns)
df.head()
```

```python
# 기본 정보 확인 1
df.info()
```

```python
# 기본 정보 확인 2
df.describe()
```

### 기존 데이터로 부터 Feature Engineering 하기
- Feature Engineering 이란, 기존에 존재하는 변수를 활용하여 새로운 정보를 추가로 생성하는 과정

```python
df["한국인 비율"] = round(df["총 한국인 수"] / df["전체 수"] * 100, 1)
df["외국인 비율"] = round(df["총 외국인 수"] / df["전체 수"] * 100, 1)
df["고령자 비율"] = round(df["총 65세 이상 고령자 수"] / df["전체 수"] * 100, 1)
df["한국인 중 남자 비율"] = round(df["총 한국인 남자 수"] / df["총 한국인 수"] * 100, 1)
df["한국인 중 여자 비율"] = round(df["총 한국인 여자 수"] / df["총 한국인 수"] * 100, 1)
df["외국인 중 남자 비율"] = round(df["총 외국인 남자 수"] / df["총 외국인 수"] * 100, 1)
df["외국인 중 여자 비율"] = round(df["총 외국인 여자 수"] / df["총 외국인 수"] * 100, 1)
```

```python
featured_df = df[
    [
        "동",
        "한국인 비율",
        "외국인 비율",
        "고령자 비율",
        "한국인 중 남자 비율",
        "한국인 중 여자 비율",
        "외국인 중 남자 비율",
        "외국인 중 여자 비율",
    ]
]
featured_df
```

```python
# 고령자 비율 순으로 정렬 후 Top 5
featured_df.sort_values(by="고령자 비율", ascending=False).head()
```

```python
# 외국인 비율 순으로 정렬 후 Top 5
featured_df.sort_values(by="외국인 비율", ascending=False).head()
```
