## pd.concat()

```python
import FinanceDataReader as fdr
import pandas as pd
```

```python
samsung_df = fdr.DataReader("005390", "2009-01-01", "2017-12-31")
print(samsung_df.shape)
samsung_df.head(2)
```

```python
kodex_df = fdr.DataReader("069500", "2016-01-01", "2017-12-31")
print(kodex_df.shape)
kodex_df.head(2)
```

```python
# axis 의 default 가 0
# 즉, 행 방향으로 데이터를 붙임
# => samsung_df 뒤에 kodex_df 를 이어 붙임!

# iloc 로 자른 이유는 정말 붙었는지 보려고
pd.concat([samsung_df, kodex_df]).iloc[2227:2232, :]
```

column, index alignment 특징은 그대로 적용됨!

```python
kodex_df[["Open", "Close"]].head(2)
```

```python
# 값이 존재하는 Open 과 Close 는 그 값을 쓰지만,
# 없는 애들은 NaN 으로 채워지는 것
pd.concat([samsung_df, kodex_df[["Open", "Close"]]]).tail(2)
```

keys, names 사용하기

```python
# 두 개의 dataframe 을 합치지만 각각에 key 를 붙여서
# 어디서 왔는지 표시하는 것!
# 즉, multi-index 를 만든다.
# multi-index: (key, 원래 자기꺼)

# iloc 로 자른 이유는 multi index 설정 보여주려고
pd.concat([samsung_df, kodex_df], keys=["samsung", "kodex"]).iloc[2227:2232, :]
```

```python
# names 도 넘겨서, keys index 에도 이름을 붙일 수 있다.

# iloc 로 자른 이유는 multi index 설정 보여주려고
pd.concat([samsung_df, kodex_df], keys=["samsung", "kodex"], names=["종목명"]).iloc[2227:2232, :]
```

axis = 1 로 했을 때는? => 열 방향으로 붙음

```python
# kodex_df 는 2016 년 부터 데이터가 있으므로,
# 없는 date(index) 에는 NaN 이 들어 감
pd.concat([samsung_df, kodex_df], axis=1).head()
```

```python
# axis 가 1 이니까 이번엔 column 이 multi-index
pd.concat([samsung_df, kodex_df], axis=1, keys=["samsung", "kodex"]).head()
```

`join` argument 를 줄 수 있음.

이 argument 는 axis 로 명시된 concat 대상 말고,

다른 애들 즉,
    axis=0 일 때는 column,
    axis=1 일 때는 index

을 어떻게 합칠 것인가에 대한 것

- inner
- outer(default)

```python
# default 'outer' join
pd.concat([samsung_df, kodex_df], keys=["삼성", "kodex"], axis=1, names=["종목명"]).head()
```

```python
# join = inner (date intersection)
pd.concat(
    [samsung_df, kodex_df],
    keys=["삼성", "kodex"],
    axis=1,
    names=["종목명"],
    join="inner",
).head()
```

```python
# concat 방향이 axis=0이니까, axis=1에 대해서 join이 적용됨
pd.concat([samsung_df.head(), kodex_df[["Close"]].head()], join="inner")
```

- 주의 : `outer` join & column 명이 서로 겹치는 게 없을 때
    => alignment 가 일치하는게 없으니 NaN으로 메꾼다!

```python
samsung_diff_col_df = samsung_df.copy()
samsung_diff_col_df.columns = ["1_" + col for col in samsung_df.columns]
samsung_diff_col_df.head()
```

```python
samsung_df.head()
```

```python
pd.concat([samsung_diff_col_df, kodex_df]).head()
```

## pd.DataFrame.join()

- 보통 index 끼리 데이터프레임을 결합
- column-index 는 `on` 인자를 사용해서 결합
- SQL의 LEFT JOIN과 유사
- 주로 index 기반 결합에 사용

```python
left = pd.DataFrame({"A": ["A0", "A1", "A2"], "B": ["B0", "B1", "B2"]}, index=["K0", "K1", "K2"])
left
```

```python
right = pd.DataFrame({"C": ["C0", "C2", "C3"], "D": ["D0", "D2", "D3"]}, index=["K0", "K2", "K3"])
right
```

```python
left.join(right)  # left join
```

```python
left.join(right, how="outer")  # 합집합!
```

```python
left = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "key": ["K0", "K1", "K0", "K1"],
    }
)
left
```

```python
right = pd.DataFrame(
    {
        "C": ["C0", "C1"],
        "D": ["D0", "D1"],
    },
    index=["K0", "K1"],
)
right
```

```python
left.join(right, on="key")
```

```python
left.join(right, on="key").set_index("key")
```

```python
left.set_index("key").join(right)
```

l_suffix, r_suffix argument
- join 할 때, 왼쪽 꺼의 column 끝에 붙일 거 -> l_suffix
             오른쪽 꺼의 column 끝에 붙일 거 -> r_suffix

```python
a = pd.DataFrame([1, 2, 3], index=["a", "b", "c"], columns=["안녕"])
a
```

```python
b = pd.DataFrame([4, 2, 6], index=["a", "c", "d"], columns=["안녕"])
b
```

```python
# a 랑 b 의 column 이 같은데, 합칠때 어디서 온 것인지 표시해주고 싶은 것!
a.join(b, lsuffix="_x", rsuffix="_y", how="inner")
```

```python
# 예전 예제를 join 으로 다시 해보기
df = pd.read_csv("../data/Small_and_Big.csv", index_col=[0])
df.head(3)
```

## pd.merge()
- DataFrame.merge() 도 쓸 수 있지만 보통 pd.merge() 를 사용함
- 특정 column을 기준으로 데이터프레임을 결합
- SQL JOIN과 매우 유사한 방식으로 동작
- 결합 방식 지정 가능: how={'inner', 'outer', 'left', 'right'}
- 여러 column을 기준으로 결합 가능

- Aligns the calling DataFrame's *column(s)* with the other DataFrame's *column(s)*
    - `left_index`, `right_index` argument도 존재하긴 함(index-index alignment시)
    - `join()`
        - 사실 내부적으로 `reset_index()` 하고 `merge()` 호출함
- Cartesian product joining
- Defaults to `inner` join
- `concat()`과 달리, index, column명이 아니라, value 값 자체를 이용한 join


- `join` 과 다른 점은 주요 차이점
    - join은 주로 index 기반, merge는 주로 column 기반 결합
    - merge가 더 유연한 결합 옵션 제공(on, how, suffixes 등)
    - merge는 복잡한 데이터 결합에 더 적합

![inner_join](https://www.w3schools.com/sql/img_inner_join.png)
![left_join](https://www.w3schools.com/sql/img_left_join.png)
![right_join](https://www.w3schools.com/sql/img_right_join.png)
![outer_join](https://www.w3schools.com/sql/img_full_outer_join.png)

```python
left = pd.DataFrame(
    {
        "key1": ["K0", "K0", "K1", "K2"],
        "key2": ["K0", "K1", "K0", "K1"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)
left
```

```python
right = pd.DataFrame(
    {
        "key1": ["K0", "K1", "K1", "K2"],
        "key2": ["K0", "K0", "K0", "K0"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
right
```

```python
left.merge(right, on=["key1", "key2"])
```

```python
left.merge(right, on=["key1", "key2"], how="outer")
```

```python
left.merge(right, on=["key1", "key2"], how="left")
```
